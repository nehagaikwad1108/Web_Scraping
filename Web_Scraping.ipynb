```python
import urllib3
import unicodedata
import requests
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
```

```python
#Code to get response from URL Server
url='http://www.bbc.com/news'
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
#Getting External Links
page = requests.get('http://www.bbc.com/news/world')
soup = BeautifulSoup(page.text, 'html.parser')
name_list = soup.find(class_='direction')
name_items = name_list.find_all('a')
for name in name_items:
   print(name.prettify())
```

```python
# Associate the link with a textual description from the website
for name in name_items:
    
    links = 'http://www.bbc.com/news/world' + name.get('href')
    names = name.text
    print(links)
    print(names)
```

```python
#Code for saving external links(urls),textual description,validity of url and last valid datetime check 
#to an excel file
import xlwt
from tempfile import TemporaryFile
from datetime import datetime

#Function to check whether link is valid or not
def get_status_code(link):
    
    
        code = requests.get(links).status_code
        print(code)
        if (code == 200):
            valid='True'
            return (valid)
        
        
bk=xlwt.Workbook()
sheet= bk.add_sheet('sheet')
sheet.write(0,0,'External Links(Urls)')
sheet.write(0,1,'Textual DEscription of Links')
sheet.write(0,2,'TimeStamp')
sheet.write(0,3,'Validity of Url')
i=1
for name in name_items:
    links='http://www.bbc.com/news/world'
    print(links)
    sheet.write(i,0,links)
    names=name.text
    print(names)
    sheet.write(i,1,names)
    isvalid= get_status_code(links)
    sheet.write(i,2,isvalid)
    time=datetime.now()
    r=time.strftime('%m/%d/%Y %H:%M:%S')
    print(r)
    sheet.write(i,3,r)
    i+=1
    
    filename="Web Scrapping Demo1233.xls"
    bk.save(filename)
    bk.save(TemporaryFile())
   # print('New File is created')
    
```
